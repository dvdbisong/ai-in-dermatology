# Rename files in folders

```python
import os
```

```python
for dirname in os.listdir("../data/"):
#     print(dirname)
    if os.path.isdir("../data/"+dirname):
#         print('Enter: ',dirname)
        for i, filename in enumerate(os.listdir("../data/"+dirname)):
            os.rename("../data/"+dirname + "/" + filename,
                      "../data/"+dirname + "/" + dirname + '-' + str(i) + ".jpg")
```
